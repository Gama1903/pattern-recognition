```python
%matplotlib inline
import os
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# 加载数据集
data_file = os.path.join('..', 'data', 'data_salmonbass.xlsx')
data = pd.read_excel(data_file)

# 初始化参数
mpl.rcParams['font.family'] = 'SimHei'

lightness_salmon = data.iloc[:74, 0]
length_salmon = data.iloc[:74, 1]
lightness_seabass = data.iloc[74:, 0]
length_seabass = data.iloc[74:, 1]

color_facecolor = '#EBEBEB'
color_salmon = '#F8766D'
color_seabass = '#00BFC4'
color_grid = '#F8F8F8'
```

```python
# 散点图
fig = plt.figure(facecolor=color_facecolor)
ax = plt.subplot(facecolor=color_facecolor)
ax.set(title='散点图', xlabel='光泽度', ylabel='长度')

plt.plot(lightness_salmon,
         length_salmon,
         color=color_salmon,
         linestyle=' ',
         marker='o',
         label='salmon')
plt.plot(lightness_seabass,
         length_seabass,
         color=color_seabass,
         linestyle=' ',
         marker='^',
         label='sea bass')

plt.legend()
plt.grid(True, color=color_grid)
```
