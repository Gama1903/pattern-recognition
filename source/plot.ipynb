```python
%matplotlib inline
import os
import pandas as pd
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 加载数据集
data_file = os.path.join('..', 'data', 'data_salmonbass.xlsx')
data_raw = pd.read_excel(data_file)

lightness_salmon = data_raw.iloc[:74, 0]
length_salmon = data_raw.iloc[:74, 1]
lightness_seabass = data_raw.iloc[74:, 0]
length_seabass = data_raw.iloc[74:, 1]
```

```python
# 加载图片
image_salmon_original = Image.open(os.path.join('..', 'image', 'salmon.png'))
image_seabass_original = Image.open(os.path.join('..', 'image', 'seabass.png'))
# 按比例缩放
scaling_ratio = 40
image_salmon_resized = image_salmon_original.resize(
    (int(image_salmon_original.size[0] / scaling_ratio),
     int(image_salmon_original.size[1] / scaling_ratio)),
    Image.Resampling.LANCZOS)
image_seabass_original.thumbnail(
    (int(image_seabass_original.size[0] / scaling_ratio),
     int(image_seabass_original.size[1] / scaling_ratio)),
    Image.Resampling.LANCZOS)
```

```python
# 初始化散点图参数
mpl.rcParams['font.family'] = 'SimHei'

color_facecolor = '#EBEBEB'
color_salmon = '#F8766D'
color_seabass = '#00BFC4'
color_grid = '#F8F8F8'
```

```python
# 散点图
fig = plt.figure(facecolor=color_facecolor)
ax = plt.subplot(facecolor=color_facecolor, xlabel='光泽度', ylabel='长度')

ax.plot(lightness_salmon,
        length_salmon,
        color=color_salmon,
        linestyle=' ',
        marker='o',
        label='salmon')
ax.plot(lightness_seabass,
        length_seabass,
        color=color_seabass,
        linestyle=' ',
        marker='^',
        label='sea bass')

# 展示图片(未实现)

plt.grid(True, color=color_grid)
plt.legend()
```
